```python
import herepy
```

```python
geocoderReverseApi = herepy.GeocoderReverseApi('Jr_STEC0z1NBe5LF5VWv8ZtYXBCx0YNi7agG3M3dqFY')
geocoderApi = herepy.GeocoderApi('Jr_STEC0z1NBe5LF5VWv8ZtYXBCx0YNi7agG3M3dqFY')
```

```python
import csv
```

```python
with open('file.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

data
```

```python
def getLatitude(address):
    response = geocoderApi.free_form(address)
    return response.Response['View'][0]['Result'][0]['Location']['DisplayPosition']['Latitude']
```

```python
def getLongitude(address):
    response = geocoderApi.free_form(address)
    return response.Response['View'][0]['Result'][0]['Location']['DisplayPosition']['Longitude']
```

```python
def getAddress(latitude, longitude):
    response = geocoderReverseApi.retrieve_addresses([latitude, longitude])
    return response.Response['View'][0]['Result'][0]['Location']['Address']['City']
```

```python
getAddress(54.21282,37.66765)
```

```python
for i in range (1,61):
    data[i].append(getLatitude(data[i][0]))
    data[i].append(getLongitude(data[i][0]))
    data[i].append(getAddress(data[i][1], data[i][2]))
data
```
